```python
from scipy.stats import f_oneway

# Test scores for each group
group_a = [70, 68, 72]
group_b = [85, 88, 82]
group_c = [80, 75, 78]


# ANOVA test
f_stat, p_value = f_oneway(group_a, group_b, group_c)
print("F-statistic:", f_stat)
print("p-value:", p_value)

# final result
if p_value < 0.05:
    print("Result: There is a significant difference between the groups.")
else:
    print("Result: No significant difference between the groups.")

```
