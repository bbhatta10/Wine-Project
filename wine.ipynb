```python
# 'Heatmap'
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(15, 8))

sns.heatmap(
    red_wine.corr(numeric_only=True),
    mask=np.triu(np.ones_like(red_wine.corr(numeric_only=True), dtype=bool)),
    annot=True,
    cmap='coolwarm',
    vmin=-1,
    vmax=1
)

plt.title('Red Wine Data Correlation Heatmap')
plt.show()

```
